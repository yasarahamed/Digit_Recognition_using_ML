```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = '/gdrive/MyDrive/ML/data/'
```

```python
data = pd.read_csv(path+'paint_mnist.csv', header=None)
data.head()
```

```python
data.shape
```

```python
x=data.drop(columns=[0])
x.insert(0,'bias',1)
x.head()
```

```python
x=np.array(x)
```

```python
y=data[0]
y.shape
```

```python
y=np.array(y)
y=np.reshape(y,(-1,1))
```

```python
x.shape, y.shape
```

```python
classes = np.unique(y).size  #finding all unique classes in the output
print(classes)
```

```python
params = x.shape[1]             
all_theta = np.zeros((classes, params))
all_theta.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=2)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def computeCost(theta, x, y):
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    first = np.multiply(-y, np.log(sigmoid(x * theta.T))) 
    second = np.multiply((1 - y), np.log(1 - sigmoid(x * theta.T)))
    
    cost = np.sum(first - second)
    cost = cost/(2*len(x)) 
                 
    return cost
```

```python
def gradient(theta, x, y):
    
    theta = np.matrix(theta)
    x = np.matrix(x)
    y = np.matrix(y)
    learning_rate = 1
        
    error = sigmoid(x * theta.T) - y
    grad = learning_rate*((x.T * error) / len(x))
    
    return grad
```

```python
from scipy.optimize import minimize
def one_vs_all(x, y, classes, new_theta):
    params = x.shape[1]
    rows = x.shape[0]
      
    for i in range(0, classes):
        theta = new_theta[i,:]
        
        y_i = np.array([1 if label == i else 0 for label in y]) 
        y_i = np.reshape(y_i, (rows, 1))
                
        # minimize the objective function  --> Gradient descent
        fmin = minimize(fun=computeCost, x0=theta, args=(x, y_i), method='TNC', jac=gradient)
        all_theta[i,:] = fmin.x
        
        print(fmin.fun)
    
    return all_theta 
```

```python
final_theta = one_vs_all(x_train,y_train, classes, all_theta)
```

```python
print('optimized_theta values = {}'.format(final_theta))
```

```python
def predict_all(x, theta): 
    x = np.matrix(x)
    theta = np.matrix(theta)

    h = sigmoid(x * theta.T)
    
    h_argmax = np.argmax(h, axis=1)
    return h_argmax, h
```

```python
y_pred, h = predict_all(x_train, final_theta)
#print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred,y_train)
print('accuracy = {} % '.format (round(accuracy,2)*100))
```

```python
t_y_pred, h = predict_all(x_test, final_theta)
```

```python
accuracy_test = accuracy_score(t_y_pred,y_test)
print('accuracy = {} % '.format (round(accuracy_test,2)*100))
```
